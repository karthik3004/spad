```python
import pandas as pd
import numpy as np
import re
import os
import glob
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
path = r'C:\Users\karthiksp\OneDrive - A STAR\Documents\SPAD\spad\expt\data\batch_2_PIN\dark_IV\T5\R42-D63'

# df = pd.DataFrame(columns=['Voltage', 'Current', 'Gate', 'Device'])
df = []
os.chdir(path)
for file in glob.glob("*.out"):
    fragmented_file = file[:-4]
    reticule, chip, device, measurement, gate, date, time = re.split('-|_',fragmented_file)
    sdf = pd.read_csv(file, delimiter=r"\s+", skiprows=3)
    sdf = sdf.iloc[: , :-3]
    sdf = sdf.rename(columns={sdf.columns[0]: 'Voltage', sdf.columns[1]: 'Current'})
    sdf['Gate'] = gate
    sdf['Gate'] = sdf['Gate'].str.slice_replace(0,4, '')
    sdf['Gate'] = sdf['Gate'].astype(int)
    sdf['Device'] = device
    sdf['Device'] = sdf['Device'].astype(int)
    sdf['Timestamp'] = date + time
    sdf['Timestamp'] = sdf['Timestamp'].astype('int64')
    df.append(sdf)
df = pd.concat(df)
```

```python
deviceID = df.Device.unique()
gates = df.Gate.unique()

for i in deviceID:
    comparision = []
    for j in gates:
        sorting = df[(df['Device']==i) & (df['Gate']==j)]
        intermediate = sorting[sorting.Timestamp == sorting.Timestamp.max()]
        comparision.append(intermediate)
    comparision = pd.concat(comparision)
    fig = px.scatter(comparision, x = 'Voltage', y = 'Current', color='Gate', log_y=True)
    fig.update_layout(title_text=f"Dark IV plot with applied gate for Device {i}", title_font_size=18, title_x=0.5, template="none", legend_title="Deadtime (ns)", margin_b=50, margin_t=60)
    fig.update_xaxes(showgrid=False, title_text="Voltage (V)", showline=True, linewidth=2, linecolor='black', mirror=True, ticks="inside", tickwidth=2, tickcolor='black', ticklen=10, title_standoff=10)
    fig.update_yaxes(title_text="Current (A)", showline=True, linewidth=2, linecolor='black', mirror=True, ticks="inside", tickwidth=2, tickcolor='black', ticklen=10, title_standoff=10)
    fig.show()
    
```
